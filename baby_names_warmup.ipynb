# Baby Names

This dataset contains information on how many babies were born with each name. Both the raw number and proportion of births for the year are included.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/zgulde/tidytuesday/master/data/2022/2022-03-22/babynames.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.year.value_counts()
```

```python
# What were the 10 most popular names the year you were born?
names_83 = df[df['year'] == 1983]
names_83.sort_values('n', ascending=False).head(10)
```

```python
# What year was your name the most popular? 2017
abdiel_df = df[df['name']=='Abdiel']
abdiel_df.max()
```

```python
names_83['rank']= names_83.n.rank(ascending=False)
names_83[names_83.name=='Abdiel']
```

```python
# Does your name increase or decrease in popularity compared to the year after the year you were born? The year before? What about 5 years before/after
abdiel_over_years = df[df['name']=='Abdiel'].sort_values('year',ascending=False)
abdiel_over_years
```

```python
# Visualize your name's popularity over time.
plt.plot(abdiel_over_years.year, abdiel_over_years.n)
```
